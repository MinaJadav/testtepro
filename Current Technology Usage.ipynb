```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data into a DataFrame
data = pd.read_csv("C:/Users/minaj/Downloads/m5_survey_data_technologies_normalised.csv")

# Get the top 10 languages worked with
top_languages = data['LanguageWorkedWith'].value_counts().head(10)

# Create a bar chart visualization of the top languages
plt.figure(figsize=(10, 5))
sns.barplot(x=top_languages.index, y=top_languages.values)
plt.xticks(rotation=45)
plt.xlabel('Programming Language')
plt.ylabel('Count')
plt.title('Top 10 Programming Languages Worked With')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("C:/Users/minaj/Downloads/m5_survey_data_technologies_normalised.csv")

# Get the Top 10 DatabaseWorkedWith
top_10_db = data['DatabaseWorkedWith'].value_counts().nlargest(10)

# Create the column chart
plt.figure(figsize=(8, 6))
ax = top_10_db.plot(kind='bar', color='purple')
ax.set_title('Top 10 DatabaseWorkedWith')
ax.set_xlabel('DatabaseWorkedWith')
ax.set_ylabel('Count')

# Show the value labels
for i, v in enumerate(top_10_db):
    ax.text(i, v + 1, str(v), ha='center')
            
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Load the data
df = pd.read_csv("C:/Users/minaj/Downloads/m5_survey_data_technologies_normalised.csv")

# Get the top 10 web frameworks
web_frameworks = df['WebFrameWorkedWith'].value_counts().head(10)

# Create a hierarchy bubble chart
fig = px.scatter(
    web_frameworks,
    x=[0] * len(web_frameworks),
    y=[0] * len(web_frameworks),
    size='WebFrameWorkedWith',
    color=web_frameworks.index,
    hover_name=web_frameworks.index,
    title='Top 10 Web Frameworks Worked With',
    width=800,
    height=600
)

# Set the axis and legend titles
fig.update_layout(
    xaxis=dict(showticklabels=False, showgrid=False, zeroline=False),
    yaxis=dict(showticklabels=False, showgrid=False, zeroline=False),
    legend=dict(title='Web Framework', orientation='h', y=1.1, x=0.5)
)

# Show the chart
fig.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
df = pd.read_csv("C:/Users/minaj/Downloads/m5_survey_data_technologies_normalised.csv")

# Get the top 10 most desired databases for next year
top_databases = df['DatabaseDesireNextYear'].value_counts().nlargest(10)

# Create a column chart
plt.figure(figsize=(10, 6))
plt.bar(top_databases.index, top_databases.values, color='#4CAF50')

# Set chart title and labels
plt.title('Top 10 Most Desired Databases for Next Year')
plt.xlabel('Database')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=90)

# Show value labels on top of the bars
for i, v in enumerate(top_databases.values):
    plt.text(i, v, str(v), ha='center', fontweight='bold')

# Display the chart
plt.show()
```

```python
# Importing necessary libraries
import pandas as pd
import plotly.express as px

# Reading the data
df = pd.read_csv("C:/Users/minaj/Downloads/m5_survey_data_technologies_normalised.csv")

# Calculating the count of each PlatformDesireNextYear value
platform_count = df['PlatformDesireNextYear'].value_counts().reset_index()
platform_count.columns = ['Platform', 'Count']

# Creating the Tree map chart
fig = px.treemap(platform_count, 
                 path=['Platform'], 
                 values='Count', 
                 color='Count', 
                 color_continuous_scale='Viridis', 
                 hover_data=['Count'],
                 labels={'Count': 'Count'},
                 color_continuous_midpoint=(platform_count['Count'].sum()/len(platform_count
['Count']))
                )

# Customizing the chart
fig.update_layout(
    title='Platform Desire Next Year',
    font=dict(size=14),
    margin=dict(t=50),
    treemapcolorway=px.colors.qualitative.Dark2,
    uniformtext=dict(minsize=12, mode='hide'),
    coloraxis=dict(colorbar=dict(title='Count', thickness=20))
)

# Displaying the chart
fig.show()                                                                              
```

```python
import pandas as pd
import plotly.express as px

# load the data
df = pd.read_csv("C:/Users/minaj/Downloads/m5_survey_data_technologies_normalised.csv")

# get Top 10 WebFrameDesireNextYear
webframe_desire = df['WebFrameDesireNextYear'].value_counts().head(10)

# create a dataframe for the chart
df_chart = pd.DataFrame({'WebFrameDesireNextYear': webframe_desire.index, 'count': webframe_desire.values})

# visualize as a Hierarchy bubble chart
fig = px.scatter(df_chart, x='WebFrameDesireNextYear', y='count', size='count', color='WebFrameDesireNextYear',
                 color_discrete_sequence=px.colors.qualitative.Pastel, text='count', height=400,
                 title='Top 10 WebFrameDesireNextYear')
fig.update_traces(marker=dict(line=dict(width=2, color='white')))
fig.update_layout(showlegend=False, xaxis_title='', yaxis_title='Count')
fig.show()
```

```python
#Demographics

import pandas as pd
import plotly.graph_objs as go
import plotly.express as px

# load data
df_demo = pd.read_csv("C:/Users/minaj/Downloads/m5_survey_data_demographics.csv")

# filter out other gender types
df_demo = df_demo[df_demo['Gender'].isin(['Man', 'Woman'])]

# count number of respondents by gender
gender_counts = df_demo['Gender'].value_counts()

# create pie chart
fig = go.Figure(data=go.Pie(labels=gender_counts.index, 
                             values=gender_counts.values, 
                             textinfo='percent',
                             hole=.4))
fig.update_traces(marker=dict(colors=['#FFC0CB', '#ADD8E6'],
                              line=dict(color='#FFFFFF', width=2)))
fig.update_layout(title='Respondents by Gender')
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Load the data
df = pd.read_csv("C:/Users/minaj/Downloads/m5_survey_data_demographics.csv")

# Filter for Man and Woman gender only
df = df[df["Gender"].isin(["Man", "Woman"])]

# Get respondent counts by country
country_counts = df["Country"].value_counts().reset_index()
country_counts.columns = ["Country", "Count"]

# Create map chart
fig = px.choropleth(country_counts, locations="Country", locationmode="country names",
                    color="Count", hover_name="Country", title="Respondent Count by Country")
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Load data
df_demographics = pd.read_csv("C:/Users/minaj/Downloads/m5_survey_data_demographics.csv")

# Filter by Gender
df_demographics = df_demographics[df_demographics['Gender'].isin(['Man', 'Woman'])]

# Group by Age and count the respondents
df_age_count = df_demographics.groupby('Age').size().reset_index(name='Respondent Count')

# Create Line chart
fig = px.line(df_age_count, x='Age', y='Respondent Count', markers=True)
fig.update_traces(mode='lines+markers')
fig.update_layout(title='Respondent Count by Age', xaxis_title='Age', yaxis_title='Respondent Count')
fig.show()
```
